```python
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns 
# импортируем библиотеки
```

```python
flowers = pd.read_csv('/carpov/iris.csv')
# считываем CSV  и назначаем переменную
```

```python
flowers.dtypes
flowers.isnull().sum()
# проверяем формат данных и смотрим количество пустых коллонок
```

```python
flowers = flowers.rename(columns={'Unnamed: 0': 'id_flower', 
                                  'sepal length': 'sepal_length', 
                                  'sepal width': 'sepal_width', 
                                  'petal length': 'petal_length', 
                                  'petal width': 'petal_width' })
# приведем наименования колонок в более удобный вид
```

```python
#  Размах в значениях находиться как разница между минимальным и максимальным значением
flowers.sepal_length.describe()
flowers.petal_length.describe()
```

```python
flowers.petal_length.describe()
```

```python
flowers.sepal_width.describe()
```

```python
flowers.petal_width.describe()
```

```python
fig, q = plt.subplots(2, 2)
sns.histplot(data=flowers, x='sepal_length', color="red", kde=True, legend='Sepal length', ax=q[0,0])
sns.histplot(data=flowers, x='petal_length', color="yellow", kde=True, legend='Petal length', ax=q[0,1])
sns.histplot(data=flowers, x='sepal_width', color="blue", kde=True, legend='Sepal width', ax=q[1,0])
sns.histplot(data=flowers, x='petal_width', color="green", kde=True, legend='Petal width', ax=q[1,1])
#  fig, q = plt.subplots(2, 2) позволяет разместить несколько графиков рядом (кол-во графиков указыватся в 
# как размер квадрата и передается в качестве аргумента к методу субплот, а также необходимо указать расположение 
#  графика в нужном квадрате ax=q[0,0] )
```

```python

sns.histplot(data=flowers, x='sepal_length', color="red", kde=True, legend='Sepal length')
sns.histplot(data=flowers, x='petal_length', color="yellow", kde=True, legend='Sepal width')
sns.histplot(data=flowers, x='sepal_width', color="blue", kde=True, legend='Petal length')
sns.histplot(data=flowers, x='petal_width', color="green", kde=True, legend='Petal width')
# Для того, чтобы размместить несколько графиков в одном поле необходимо перечислить все аргументы как в коде выше)
```

```python
sns.violinplot(data=flowers["petal length"])
```

```python
sns.pairplot(flowers, hue="species")
```
